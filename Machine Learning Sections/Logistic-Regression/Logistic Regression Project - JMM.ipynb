# Logistic Regression Project 

In this project we will be working with a fake advertising data set, indicating whether or not a particular internet user clicked on an Advertisement. We will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

# Logistic Regression

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

**BEFORE splitting the data, let us clean up the data frame a little.**

First check how many unique values there are in the **City** column.

```python
ad_data['City'].nunique()
```

There are 969 unique cities, which is nearly the same number as the number of observations in the data set (1000). So **City** is unlikely to have good predictive power. So, I will leave it out of the modeling.

```python
ad_data.drop('City',axis=1,inplace=True)
```

Now let us check the **Country** column.

```python
ad_data['Country'].nunique()
```

**Country** has 237 unique values. Maybe we can retain this as a predictor, but I am skeptical. Let us drop this columns.

```python
ad_data.drop('Country',axis=1,inplace=True)
```

Now let us check the **Ad Topic Line** column.

```python
ad_data['Ad Topic Line'].nunique()
```

**Ad Topic Line** has 1000 unique values, too many to be useful. I will leave it out of the modeling.

```python
ad_data.drop('Ad Topic Line',axis=1,inplace=True)
```

Finally, consdier the **Timestamp** column. This is a character string which we could potentially convert to a number. But let us just drop this column too.

```python
ad_data.drop('Timestamp',axis=1,inplace=True)
```

Let us see what we have so far.

```python
ad_data.info()
```

```python
ad_data.head()
```

OK, all columns are numeric now, so we can proceed.

** Split the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(ad_data.drop('Clicked on Ad',axis=1), 
                                                    ad_data['Clicked on Ad'], test_size=0.30, 
                                                    random_state=101)
```

** Train and fit a logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
predictions = logmodel.predict(X_test)
```

** Create a classification report for the model.**

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
